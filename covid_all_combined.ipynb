<h2>Load the Data</h2>

<h2>Transform Dataframes</h2>

<h3>Fat Supply</h3>

```python
## FAT SUPPY ##

# Create a filtered dataframe from specific columns
fat_supply_cols = ["Country", "Animal Products", "Cereals - Excluding Beer", "Eggs", "Fruits - Excluding Wine", \
             "Meat", "Milk - Excluding Butter", "Fish, Seafood", "Vegetables", "Obesity", "Undernourished", "Confirmed", "Deaths", "Recovered", "Population"]

# food_cols = ["Country", "Animal_Products", "Cereals", "Eggs", "Fruits", "Meat", "Milk", "Fish_Seafood", "Vegetables"]
fat_supply_transformed= fat_supply_df[fat_supply_cols].copy()

# Rename the column headers
fat_supply_transformed = fat_supply_transformed.rename(columns={'Country': 'country', 
                                                    'Animal Products':'animal_products_fat',
                                                    'Cereals - Excluding Beer':'cereals_fat',
                                                    'Eggs':'eggs_fat',
                                                    'Fruits - Excluding Wine': 'fruits_fat',
                                                    'Meat':'meat_fat',
                                                    'Milk - Excluding Butter':'milk_fat',
                                                    'Fish, Seafood': 'fish_seafood_fat', 
                                                    'Vegetables':'vegetables_fat',
                                                    'Obesity': 'obesity',
                                                    'Undernourished': 'undernourished',
                                                    'Confirmed': 'confirmed_cases',
                                                    'Deaths': 'deaths',
                                                    'Recovered': 'recovered',
                                                    'Population': 'population'})
# Clean the data by dropping duplicates and setting the index
fat_supply_transformed.drop_duplicates("country", inplace=True)
fat_supply_transformed.set_index("country", inplace=True)

fat_supply_transformed
```

<h3>Food Quantity (kg)</h3>

```python
## FOOD QUANTITY ##

# Create a filtered dataframe from specific columns
food_cols = ["Country", "Animal Products", "Cereals - Excluding Beer", "Eggs", "Fruits - Excluding Wine", \
             "Meat", "Milk - Excluding Butter", "Fish, Seafood", "Vegetables", "Obesity", "Undernourished", "Confirmed", "Deaths", "Recovered", "Population"]

# food_cols = ["Country", "Animal_Products", "Cereals", "Eggs", "Fruits", "Meat", "Milk", "Fish_Seafood", "Vegetables"]
food_transformed= food_quantity_df[food_cols].copy()

# Rename the column headers
food_transformed = food_transformed.rename(columns={'Country': 'country', 
                                                    'Animal Products':'animal_products_quantity',
                                                    'Cereals - Excluding Beer':'cereals_quantity',
                                                    'Eggs':'eggs_quantity',
                                                    'Fruits - Excluding Wine': 'fruits_quantity',
                                                    'Meat':'meat_quantity',
                                                    'Milk - Excluding Butter':'milk_quantity',
                                                    'Fish, Seafood': 'fish_seafood_quantity', 
                                                    'Vegetables':'vegetables_quantity',
                                                    'Obesity': 'obesity',
                                                    'Undernourished': 'undernourished',
                                                    'Confirmed': 'confirmed_cases',
                                                    'Deaths': 'deaths',
                                                    'Recovered': 'recovered',
                                                    'Population': 'population'})
# Clean the data by dropping duplicates and setting the index
food_transformed.drop_duplicates("country", inplace=True)
food_transformed.set_index("country", inplace=True)


food_transformed
```

<h3>KCal Supply</h3>

```python
## KCAL SUPPY ##

# Create a filtered dataframe from specific columns
kcal_supply_cols = ["Country", "Animal Products", "Cereals - Excluding Beer", "Eggs", "Fruits - Excluding Wine", \
             "Meat", "Milk - Excluding Butter", "Fish, Seafood", "Vegetables", "Obesity", "Undernourished", "Confirmed", "Deaths", "Recovered", "Population"]

# food_cols = ["Country", "Animal_Products", "Cereals", "Eggs", "Fruits", "Meat", "Milk", "Fish_Seafood", "Vegetables"]
kcal_supply_transformed= kcal_supply_df[kcal_supply_cols].copy()

# Rename the column headers
kcal_supply_transformed = kcal_supply_transformed.rename(columns={'Country': 'country', 
                                                    'Animal Products':'animal_products_kcal',
                                                    'Cereals - Excluding Beer':'cereals_kcal',
                                                    'Eggs':'eggs_kcal',
                                                    'Fruits - Excluding Wine': 'fruits_kcal',
                                                    'Meat':'meat_kcal',
                                                    'Milk - Excluding Butter':'milk_kcal',
                                                    'Fish, Seafood': 'fish_seafood_kcal', 
                                                    'Vegetables':'vegetables_kcal',
                                                    'Obesity': 'obesity',
                                                    'Undernourished': 'undernourished',
                                                    'Confirmed': 'confirmed_cases',
                                                    'Deaths': 'deaths',
                                                    'Recovered': 'recovered',
                                                    'Population': 'population'})
# Clean the data by dropping duplicates and setting the index
kcal_supply_transformed.drop_duplicates("country", inplace=True)
kcal_supply_transformed.set_index("country", inplace=True)


kcal_supply_transformed
```

<h3>Protein Supply</h3>

```python
## PROTEIN SUPPLY ##

# Create a filtered dataframe from specific columns
protein_supply_cols = ["Country", "Animal Products", "Cereals - Excluding Beer", "Eggs", "Fruits - Excluding Wine", \
             "Meat", "Milk - Excluding Butter", "Fish, Seafood", "Vegetables", "Obesity", "Undernourished", "Confirmed", "Deaths", "Recovered", "Population"]

# food_cols = ["Country", "Animal_Products", "Cereals", "Eggs", "Fruits", "Meat", "Milk", "Fish_Seafood", "Vegetables"]
protein_supply_transformed= protein_supply_df[protein_supply_cols].copy()

# Rename the column headers
protein_supply_transformed = protein_supply_transformed.rename(columns={'Country': 'country', 
                                                    'Animal Products':'animal_products_protein',
                                                    'Cereals - Excluding Beer':'cereals_protein',
                                                    'Eggs':'eggs_protein',
                                                    'Fruits - Excluding Wine': 'fruits_protein',
                                                    'Meat':'meat_protein',
                                                    'Milk - Excluding Butter':'milk_protein',
                                                    'Fish, Seafood': 'fish_seafood_protein', 
                                                    'Vegetables':'vegetables_protein',
                                                    'Obesity': 'obesity',
                                                    'Undernourished': 'undernourished',
                                                    'Confirmed': 'confirmed_cases',
                                                    'Deaths': 'deaths',
                                                    'Recovered': 'recovered',
                                                    'Population': 'population'})
# Clean the data by dropping duplicates and setting the index
protein_supply_transformed.drop_duplicates("country", inplace=True)
protein_supply_transformed.set_index("country", inplace=True)

protein_supply_transformed
```

<h3>Merge dataframes into one</h3>

```python
dfs = [food_transformed, kcal_supply_transformed, fat_supply_transformed, protein_supply_transformed]
nan_value = 0

##BECAUSE EACH DATASET HAS THE SAME DATA IN OBESITY, UNDERNOURISHED, CONFIRMED_CASES, DEATHS, RECOVERED, AND POPULATION THEY
##CAN BE MERGED INTO ONE LARGE DATAFRAME

covid_diet = functools.reduce(lambda left, right: pd.merge(left,right,on=['country', 'obesity', 'undernourished', 'confirmed_cases', 'deaths', 'recovered', 'population']), dfs)
covid_diet
```
