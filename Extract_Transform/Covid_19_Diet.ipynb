```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
Fat_Supply_Quantity = "../Resources/Fat_Supply_Quantity_Data.csv"
Fat_Supply_Quantity_df = pd.read_csv(Fat_Supply_Quantity)
Fat_Supply_Quantity_df.head()
```

```python
Food_Supply_Quantity = "../Resources/Food_Supply_Quantity_kg_Data.csv"
Food_Supply_Quantity_df = pd.read_csv(Food_Supply_Quantity)
Food_Supply_Quantity_df.head()
```

```python
Food_Supply_Kcal = "../Resources/Food_Supply_kcal_Data.csv"
Food_Supply_Kcal_df = pd.read_csv(Food_Supply_Kcal)
Food_Supply_Kcal_df.head()
```

```python
Protein_Supply_Quantity = "../Resources/Protein_Supply_Quantity_Data.csv"
Protein_Supply_Quantity_df = pd.read_csv(Protein_Supply_Quantity)

Protein_Supply_Quantity_df.head()
```

## Create filtered Dataframes

```python
#Create a filtered dataframe from specific columns
Food_Supply_Quantity_cols = ['Country', 'Animal Products', 'Cereals - Excluding Beer',
                             'Eggs', 'Fruits - Excluding Wine', 'Meat', 'Milk - Excluding Butter', 'Fish, Seafood', 'Vegetables']
Food_Supply_Quantity_transformed= Food_Supply_Quantity_df[Food_Supply_Quantity_cols].copy()

#Rename column headers 
Food_Supply_Quantity_transformed = Food_Supply_Quantity_transformed.rename(columns={'Country': 'country', 
                                                                                    'Animal Products':'animal_products',
                                                                                    'Cereals - Excluding Beer':'cereals',
                                                                                    'Eggs':'eggs',
                                                                                    'Fruits - Excluding Wine': 'fruits',
                                                                                    'Meat':'meat',
                                                                                    'Milk - Excluding Butter':'milk',
                                                                                    'Fish, Seafood': 'fish_seafood', 
                                                                                    'Vegetables':'vegetables'})


#Clean the data by dropping duplicates and setting the index
Food_Supply_Quantity_transformed.drop_duplicates("country", inplace=True)
Food_Supply_Quantity_transformed.set_index("country", inplace=True)

Food_Supply_Quantity_transformed.insert(0,'type', 2)

Food_Supply_Quantity_transformed
```

```python
#Create a filtered dataframe from specific columns
Fat_Supply_Quantity_cols = ['Country', 'Animal Products', 'Cereals - Excluding Beer',
                             'Eggs', 'Fruits - Excluding Wine', 'Meat', 'Milk - Excluding Butter', 'Fish, Seafood', 'Vegetables']
Fat_Supply_Quantity_transformed= Fat_Supply_Quantity_df[Fat_Supply_Quantity_cols].copy()

#Rename column headers 
Fat_Supply_Quantity_transformed = Fat_Supply_Quantity_transformed.rename(columns={'Country': 'country', 
                                                                                    'Animal Products':'animal_products',
                                                                                    'Cereals - Excluding Beer':'cereals',
                                                                                    'Eggs':'eggs',
                                                                                    'Fruits - Excluding Wine': 'fruits',
                                                                                    'Meat':'meat',
                                                                                    'Milk - Excluding Butter':'milk',
                                                                                    'Fish, Seafood': 'fish_seafood', 
                                                                                    'Vegetables':'vegetables'})


#Clean the data by dropping duplicates and setting the index
Fat_Supply_Quantity_transformed.drop_duplicates("country", inplace=True)
Fat_Supply_Quantity_transformed.set_index("country", inplace=True)

Fat_Supply_Quantity_transformed.insert(0,'type', 1)

Fat_Supply_Quantity_transformed
```

```python
#Create a filtered dataframe from specific columns
Food_Supply_Kcal_cols = ['Country', 'Animal Products', 'Cereals - Excluding Beer',
                             'Eggs', 'Fruits - Excluding Wine', 'Meat', 'Milk - Excluding Butter', 'Fish, Seafood', 'Vegetables']
Food_Supply_Kcal_transformed= Food_Supply_Kcal_df[Food_Supply_Kcal_cols].copy()

#Rename column headers 
Food_Supply_Kcal_transformed = Food_Supply_Kcal_transformed.rename(columns={'Country': 'country', 
                                                                                    'Animal Products':'animal_products',
                                                                                    'Cereals - Excluding Beer':'cereals',
                                                                                    'Eggs':'eggs',
                                                                                    'Fruits - Excluding Wine': 'fruits',
                                                                                    'Meat':'meat',
                                                                                    'Milk - Excluding Butter':'milk',
                                                                                    'Fish, Seafood': 'fish_seafood', 
                                                                                    'Vegetables':'vegetables'})


#Clean the data by dropping duplicates and setting the index
Food_Supply_Kcal_transformed.drop_duplicates("country", inplace=True)
Food_Supply_Kcal_transformed.set_index("country", inplace=True)

Food_Supply_Kcal_transformed.insert(0,'type', 3)

Food_Supply_Kcal_transformed
```

```python
#Create a filtered dataframe from specific columns
Protein_Supply_Quantity_cols = ['Country', 'Animal Products', 'Cereals - Excluding Beer',
                             'Eggs', 'Fruits - Excluding Wine', 'Meat', 'Milk - Excluding Butter', 'Fish, Seafood', 'Vegetables']
Protein_Supply_Quantity_transformed= Protein_Supply_Quantity_df[Protein_Supply_Quantity_cols].copy()

#Rename column headers 
Protein_Supply_Quantity_transformed = Protein_Supply_Quantity_transformed.rename(columns={'Country': 'country', 
                                                                                    'Animal Products':'animal_products',
                                                                                    'Cereals - Excluding Beer':'cereals',
                                                                                    'Eggs':'eggs',
                                                                                    'Fruits - Excluding Wine': 'fruits',
                                                                                    'Meat':'meat',
                                                                                    'Milk - Excluding Butter':'milk',
                                                                                    'Fish, Seafood': 'fish_seafood', 
                                                                                    'Vegetables':'vegetables'})

#Clean the data by dropping duplicates and setting the index
Protein_Supply_Quantity_transformed.drop_duplicates("country", inplace=True)
Protein_Supply_Quantity_transformed.set_index("country", inplace=True)

Protein_Supply_Quantity_transformed.insert(0,'type', 4)

Protein_Supply_Quantity_transformed
```

```python
#Create a filtered dataframe from specific columns
covid_cols = ['Country', 'Obesity', 'Undernourished',
                             'Confirmed', 'Deaths', 'Recovered', 'Active', 'Population'] 
covid_cols_transformed = covid_cols_df[covid_cols].copy()

#Rename column headers 
covid_cols_transformed = covid_cols_transformed.rename(columns={'Country': 'country', 
                                                                'Obesity': 'obesity',
                                                                'Undernourished': 'undernourished',
                                                                'Confirmed':'confirmed',
                                                                'Deaths':'deaths',
                                                                'Recovered':'recovered',
                                                                'Active':'active',
                                                                'Population':'population'})

#Clean the data by dropping duplicates and setting the index
covid_cols_transformed.drop_duplicates("country", inplace=True)
covid_cols_transformed.set_index("country", inplace=True)

covid_cols_transformed
```

## Concatenate  dataframes together 
Create one larger dataframe with all values from Fat Supply, Food Supply, Kcal, and Protein Suppy.

```python
dfs = [Fat_Supply_Quantity_transformed, Food_Supply_Quantity_transformed, Food_Supply_Kcal_transformed, Protein_Supply_Quantity_transformed]
nan_value = 0

Covid_19_Diet = pd.concat(dfs).fillna(nan_value)

Covid_19_Diet
```

```python
connection_string = "postgres:Kemb@2010@localhost:5432/ETL_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

## Load DataFrames into database

```python
Fat_Supply_Quantity_transformed.to_sql(name='fat_supply_quantity', con=engine, if_exists='append', index=True)
```

```python
Food_Supply_Quantity_transformed.to_sql(name='food_supply_quantity', con=engine, if_exists='append', index=True)
```

```python
Food_Supply_Kcal_transformed.to_sql(name='food_supply_kcal', con=engine, if_exists='append', index=True)
```

```python
Protein_Supply_Quantity_transformed.to_sql(name='protein_supply_quantity', con=engine, if_exists='append', index=True)
```

```python
Covid_19_Diet.to_sql(name='covid_19_diet', con=engine, if_exists='append', index=True)
```

```python
covid_cols_transformed.to_sql(name='covid_data', con=engine, if_exists='append', index=True)
```
